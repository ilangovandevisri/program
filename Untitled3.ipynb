```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

num = int(input())
rev = int(str(num)[::-1])

if is_prime(num) and is_prime(rev):
    print("Both are prime")
else:
    print("Not prime")
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

num = input()
rotations = []

for i in range(len(num)):
    rotations.append(int(num[i:] + num[:i]))

if all(is_prime(x) for x in rotations):
    print("Circular Prime")
else:
    print("Not Circular Prime")
```

```python
def fact(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    return f

num = int(input())
temp = num
s = 0

while temp > 0:
    d = temp % 10
    s += fact(d)
    temp //= 10

if s == num:
    print("Peterson Number")
else:
    print("Not Peterson Number")
```

```python
mat = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

top, bottom = 0, 2
left, right = 0, 2

while top <= bottom and left <= right:
    for i in range(left, right+1):
        print(mat[top][i], end=" ")
    top += 1

    for i in range(top, bottom+1):
        print(mat[i][right], end=" ")
    right -= 1

    for i in range(right, left-1, -1):
        print(mat[bottom][i], end=" ")
    bottom -= 1

    for i in range(bottom, top-1, -1):
        print(mat[i][left], end=" ")
    left += 1
```

```python
def is_pal(s):
    return s == s[::-1]

s = input()

i, j = 0, len(s)-1

while i < j:
    if s[i] != s[j]:
        s1 = s[:i] + s[i+1:]
        s2 = s[:j] + s[j+1:]
        if is_pal(s1) or is_pal(s2):
            print("YES")
        else:
            print("NO")
        break
    i += 1
    j -= 1
else:
    print("YES")
```

```python
import math

x1,y1 = map(int, input("Enter x1,y1 (e.g., 1,2): ").split(','))
x2,y2 = map(int, input("Enter x2,y2 (e.g., 3,4): ").split(','))

dist = math.sqrt((x2-x1)**2 + (y2-y1)**2)

if dist < 10:
    print("Collision Risk")
else:
    print("Safe")
```

```python
while True:
    try:
        n = int(input("Enter the number of students: "))
        if n <= 0:
            print("Enter a positive number")
            continue
        break
    except ValueError:
        print("Please enter a valid integer")

students = []
topper = ""
top_avg = -1.0

for i in range(n):
    student_input_line = input(f"Enter name and 5 marks for student {i+1}: ").strip()

    if not student_input_line:
        print("Empty input skipped")
        continue

    data = student_input_line.split()

    if len(data) != 6:
        print("Invalid input: enter name and exactly 5 marks")
        continue

    name = data[0]

    try:
        marks = list(map(int, data[1:6]))
    except ValueError:
        print("Marks must be integers only")
        continue

    avg = sum(marks) / 5
    students.append((name, avg))

    if avg > top_avg:
        top_avg = avg
        topper = name

print("\nTopper:", topper)

eligible_found = False

for name, avg in students:
    if avg >= 90:
        if not eligible_found:
            print("Scholarship Eligible:")
            eligible_found = True
        print(name)

if not eligible_found:
    print("No scholarship eligible students")
```
