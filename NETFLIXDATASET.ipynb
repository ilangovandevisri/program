```python
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
print(df[['duration']].head(10))
print(df.isnull().sum())
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
print(df[['rating']].head(10))
print(df.isnull().sum())
```

```python
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
print(df[['date_added']].head(10))
print(df.isnull().sum())
```

```python
df['director'] = df['director'].fillna('Unknown')
print(df[['director']].head(10))
print(df.isnull().sum())
```

```python
df['cast'] = df['cast'].fillna('Unknown')
print(df[['cast']].head(10))
print(df.isnull().sum())
```

```python
df['country'] = df['country'].fillna('Unknown')
print(df[['country']].head(10))
print(df.isnull().sum())
```

```python
df['description'] = df['description'].fillna('Unknown')
print(df[['description']].head(10))
print(df.isnull().sum())
```

```python
df['type'] = df['type'].fillna('Unknown')
print(df[['type']].head(10))
print(df.isnull().sum())
```

```python
df['release_year'] = df['release_year'].fillna(df['release_year'].mode()[0])
print(df[['release_year']].head(10))
print(df.isnull().sum())
```

```python
df['title'] = df['title'].fillna(df['title'].mode()[0])
print(df[['title']].head(10))
print(df.isnull().sum())
```

```python
df['listed_in']=df['listed_in'].fillna(df['listed_in'].mode()[0])
print(df[['listed_in']].head(10))
print(df.isnull().sum())
```

```python
df['show_id']=df['show_id'].fillna(df['show_id'].mode()[0])
print(df[['show_id']].head(10))
print(df.isnull().sum())
```

```python
df.drop('rating',axis=1,inplace=True)
print(df)
```

```python
df.add_prefix('netflix_')
```
