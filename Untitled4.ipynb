```python
user_input = input("Enter elements : ")

nums_str = user_input.split(',')
nums = [int(x.strip()) for x in nums_str]

n = len(nums)

largest = float('-inf')
second_largest = float('-inf')

for num in nums:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print("Second Largest =", second_largest)
```

```python
arr = [1, 2, 3, 4, 5]
k = 2

k = k % len(arr)

result = arr[-k:] + arr[:-k]

print(result)
```

```python
arr = [7, 2, 5, 4, 9, 8]

even = []
odd = []

for num in arr:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

result = even + odd

print(result)
```

```python
def nested_sum(lst):
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += nested_sum(item)
        else:
            total += item

    return total

arr = [1, [2, 3], [4, [5, 6]]]

print(nested_sum(arr))
```

```python
for num in range(1, 101):

    s = str(num)

    digit_sum = 0
    digit_product = 1

    for digit in s:
        digit_sum += int(digit)
        digit_product *= int(digit)

    if digit_sum == digit_product:
        print(num, end=" ")
```

```python
arr = [2, 7, 11, 15, 3, 6]
target = 9

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] + arr[j] == target:
            print((arr[i], arr[j]))
```
